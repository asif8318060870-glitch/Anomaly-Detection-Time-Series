 Q1. What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

- Anomaly Detection is the process of identifying data points or patterns that are different from normal behavior. These unusual observations are called **anomalies** or **outliers**. It is widely used in fraud detection, network security, healthcare, manufacturing, and predictive maintenance.

1. Point Anomaly

A point anomaly is a single data point that is very different from the rest of the data.

Example:
In a bank transaction dataset, most transactions are between ₹500 and ₹20,000, but one transaction of ₹15,00,000 appears suddenly. This transaction is a point anomaly.
# 2. Contextual Anomaly

A contextual anomaly is a data point that is considered abnormal only in a specific context, such as time, location, or season.

**Example:**
A temperature of **35°C** is normal during summer but would be unusual during winter in the same city. Therefore, it is a contextual anomaly.

3. Collective Anomaly

A collective anomaly occurs when a group of related data points behaves abnormally, even though individual points may appear normal.

Example:
In network traffic monitoring, a sudden sequence of many login attempts from different IP addresses within a few minutes may indicate a cyberattack. Each login attempt may look normal individually, but together they form a collective anomaly.

In summary, anomaly detection helps identify unusual patterns in data and supports better decision-making by detecting errors, fraud, or unexpected events.

Q2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

- Isolation Forest

Isolation Forest detects anomalies by randomly separating data points. It works well with large and high-dimensional datasets and is commonly used for fraud detection and network security. Its main advantage is that it is fast and scalable.

DBSCAN :

DBSCAN is a density-based clustering algorithm. It groups closely packed data points and treats points in low-density areas as anomalies. It is useful for spatial data and datasets with irregular cluster shapes, but its performance depends on the correct choice of parameters.

Local Outlier Factor (LOF):

Local Outlier Factor detects anomalies by comparing the local density of a point with the density of its neighboring points. It is effective for identifying local outliers but can be slower on large datasets.

Comparison:
Isolation Forest is best for large datasets, DBSCAN is suitable for density-based clustering and noise detection, while LOF is ideal for finding local anomalies in datasets with varying densities.

Q3. What are the key components of a Time Series? Explain each with one example.



- A time series is data collected over time. It has four main components:

 1. Trend

It shows the long-term increase or decrease in data.

Example:Online shopping sales increasing every year.

2. Seasonality

It is a pattern that repeats at regular time intervals.

Example: Ice cream sales increase every summer.

 3. Cyclical Variation

These are changes that happen over long periods due to business or economic cycles.

Example: Car sales decrease during a recession.

 4. Irregular Variation

These are unexpected random changes caused by unusual events.

Example: A sudden drop in air travel during the COVID-19 pandemic.

These components help in understanding and forecasting time series data.

## Q5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

- **Answer:**

**AR (AutoRegressive)** uses past values of a time series to predict future values. It is mainly used for stationary data where previous observations influence future observations.

**MA (Moving Average)** uses past forecast errors to make predictions. It is useful when prediction errors from previous time steps affect future values.

**ARIMA (AutoRegressive Integrated Moving Average)** combines AR and MA models with differencing to handle non-stationary data. It is widely used for general time series forecasting.

**SARIMA (Seasonal ARIMA)** is an extension of ARIMA that includes seasonal patterns in the data. It is suitable for datasets with regular seasonal variations, such as monthly sales or airline passengers.

**SARIMAX (Seasonal ARIMA with Exogenous Variables)** further extends SARIMA by including external factors, such as weather or holidays, that may influence the target variable. It is useful when both seasonality and external variables affect forecasting results.

In summary, AR and MA are basic forecasting models, ARIMA handles non-stationary data, SARIMA adds seasonal components, and SARIMAX also considers external variables for more accurate predictions.

```python

"""
Q8. Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.

Answer:
"""

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

url = "https://drive.google.com/uc?export=download&id=1eu6DQtqaH2qmJZU_RU5tQe5hhYejjXW3"

data = pd.read_csv(url)

data["Month"] = pd.to_datetime(data["Month"])
data.set_index("Month", inplace=True)

model = SARIMAX(
    data["#Passengers"],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12)
)

result = model.fit()

forecast = result.forecast(steps=12)

plt.figure(figsize=(10, 5))
plt.plot(data["#Passengers"], label="Original")
plt.plot(forecast.index, forecast, label="Forecast")
plt.legend()
plt.title("SARIMA Forecast")
plt.show()

```

```python

"""
Q9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.

Answer:
"""

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

data = pd.read_csv('/content/NYC_taxi_fare_data.csv')

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
data["Anomaly"] = lof.fit_predict(data[["fare_amount", "trip_distance"]])

plt.figure(figsize=(8, 6))
plt.scatter(
    data["fare_amount"],
    data["trip_distance"],
    c=data["Anomaly"]
)
plt.xlabel("Fare Amount")
plt.ylabel("Trip Distance")
plt.title("LOF Anomaly Detection")
plt.show()
```
